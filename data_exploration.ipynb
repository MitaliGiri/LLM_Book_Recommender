```python
# Importing dataset from Kaggle and creating path for it

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
#Importing pandas library

import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
# Visualizing our data (Use column statistics (detailed) for better visualization)

books
```

```python
# importing libraries for data visualisation

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating heatmap for missing values in our data

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False,  ax= ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
# importing numpy library

import numpy as np
```

```python
# adding missing description column and age of book column

books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['age_of_books'] = 2025 - books['published_year']
```

```python
# taking columns we need to create heatmap

columns_of_interest = ['num_pages', 'age_of_books', 'missing_description', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method = 'spearman')
# Here we are using spearman method instead of pearsons as pearsons is better suited for continuous values than binary values and here we are using binary values (books['missing_description']

sns.set_theme(style='white')
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'label': 'Spearman Correlation'})
heatmap.set_title('Correlation Heatmap')
plt.show()
```

```python
# checking number of columns that have these values missing
#books_missing = books[~(books['description'].isna()) &
#                     ~(books['num_pages'].isna()) &
#                     ~(books['average_rating'].isna())&
#                     ~(books['published_year'].isna())]
#

# removing columns with missing books

books_missing = books[~(books['description'].isna()) &
                      ~(books['num_pages'].isna()) &
                      ~(books['average_rating'].isna())&
                      ~(books['published_year'].isna())]
```

```python
# getting distribution of categories (we have too many unnecessary ones)

books_missing['categories'].value_counts().reset_index().sort_values('count', ascending = False)

# use chart view for better visualisation of data, in its settings change y-axis --> count and type of chart --> bar
# distribution here is very uneven and we have a long tail problem
```

```python
# getting the number of words in the description

books_missing['words_in_description'] = books_missing['description'].str.split().str.len()
```

```python
# creating dataframe of books with description that has more than 25 words

books_missing_25_words = books_missing[books_missing['words_in_description'] >= 25]
```

```python
books_missing_25_words
```

```python
# Creating new column to combine title and subtitle with ':', if subtitle present

books_missing_25_words['title_and_subtitle'] = (
        np.where(books_missing_25_words['subtitle'].isna(), books_missing_25_words['title'],
                 books_missing_25_words[['title', 'subtitle']].astype(str).agg(': '.join, axis = 1))
)
```

```python
books_missing_25_words
```

```python
# Creating new column with isbn13 and description together

books_missing_25_words['tagged_description'] = books_missing_25_words[['isbn13', 'description']].astype(str).agg(' '.join, axis = 1)
```

```python
books_missing_25_words
```

```python
# Dropping unwanted columns and saving cleaned one as csv file

(
    books_missing_25_words
    .drop(['subtitle', 'missing_description', 'age_of_books', 'words_in_description'], axis = 1)
    .to_csv('books_cleaned.csv', index = False)
)
```
